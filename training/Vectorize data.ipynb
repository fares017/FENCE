```python
#Importing packages
import pandas as pd
import warnings

warnings.filterwarnings('ignore')
```

```python
import pandas as pd
from numpy import argmax
from sklearn.model_selection import train_test_split
```

```python

binetflow_file_path = 'capture20110810.binetflow'

# Read the binary flow file into a pandas DataFrame
df = pd.read_csv(binetflow_file_path, delimiter=',')
df.head()
```

```python
df.head()
```

```python
#Renaming the columns using their proper names from the previous dataset
df.rename(columns = {list(df)[1]:'Duration',list(df)[2]:'Protocol',list(df)[3]:'Source_IP',list(df)[4]:'Source_Port',
                    list(df)[6]:'Destination_IP',list(df)[7]:'Destination_Port',list(df)[8]:'Flags',list(df)[9]:'Source_Type_of_Service',
                    list(df)[10]:'Dest_Type_of_Service',list(df)[11]:'Packets',list(df)[12]:'Total_Bytes',list(df)[13]:'Source_Bytes'}, inplace = True)
```

```python
df.head()
```

```python
df.Dir.unique()
```

```python
direction = ['   ->','  <->','  <-']
df = df.loc[df['Dir'].isin(direction)]
```

```python
df #Now 2.813Million rows - Only about 10,000 values removed from 2.824Million rows
```

```python
df.replace(['   ->','  <->','  <-'],  
                            ['outgoing','two-way','incoming'],inplace=True)
# df.replace({'Dir':{'   ->': 'outgoing', '  <->': 'two-way', '  <-': 'incoming'}})
```

```python
df.Dir.unique()
```

```python
df.rename(columns = {list(df)[4]:'Direction'}, inplace = True)
```

```python
df
```

```python
df.Label.unique()
```

```python
df['Label'] = df['Label'].str.replace(r'flow=', '')
df['Label'] = df['Label'].str.replace(r'From-', '')
df['Label'] = df['Label'].str.replace(r'To-', '')
df['Label'] = df['Label'].str.extract(r'(\w+)')
df.Label.unique()
```

```python
df.dropna(inplace=True)
df #dropping all null values removes 200k records = 2.613Million rows
```

```python
df.drop(columns=['StartTime', 'Dir', 'Flags'], inplace=True)
```

```python
df['Label'].value_counts()
```

```python
# Convert 'Botnet' to 1 and any other value to 0 in the 'Label' column
df['Label'] = df['Label'].apply(lambda x: 1 if x == 'Botnet' else 0)
```

```python
df['Label'].value_counts()
```

```python
df
```

```python
df['Protocol'].value_counts()
```

```python
# One-hot encode the 'Protocol' column
protocol_encoded = pd.get_dummies(df['Protocol'], prefix='Protocol')

# Manually convert one-hot encoded columns to binary values
protocol_encoded_binary = protocol_encoded.applymap(lambda x: 1 if x == 1 else 0)

# Concatenate the one-hot encoded columns with the original DataFrame
df_encoded = pd.concat([df, protocol_encoded_binary], axis=1)

# Drop the original 'Protocol' column
df_encoded.drop(columns=['Protocol'], inplace=True)
```

```python
df_encoded
```

```python
df_encoded = pd.concat([df_encoded.drop(columns='Label'), df_encoded['Label']], axis=1)
```

```python
df_encoded
```

```python
df_encoded.columns
```

# Split the data into training and testing sets


```python
train_df, test_df = train_test_split(df_encoded, test_size=0.2, random_state=42)
```

```python
# Save training and testing sets to files
train_df.to_csv('vectorized_train_data.csv', index=False)
test_df.to_csv('vectorized_test_data.csv', index=False)
```
